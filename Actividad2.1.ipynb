```python
#Importamos librer√≠as requeridas
import pandas as pd
import missingno as msno
```

```python
# Carga el archivo CSV en un DataFrame
ruta_archivo = "ventas_totales.csv"
df = pd.read_csv(ruta_archivo)

# Calcula la cantidad de valores nulos por columna
valores_nulos_por_columna = df.isnull().sum()

# Imprime los resultados
print("Valores nulos por columna:")
print(valores_nulos_por_columna)
```

```python
df['salon_ventas'] = df['salon_ventas'].fillna(round(df['salon_ventas'].mean(),1))
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df['tarjetas_debito'] = df['tarjetas_debito'].fillna(round(df['tarjetas_debito'].median(),1))
valores_nulos=df.isnull().sum()
valores_nulos
```

- Eliminaci√≥n de filas o columnas con valores nulos:
Cu√°ndo usarlo: Esta estrategia se utiliza cuando la cantidad de valores nulos es peque√±a y no afecta significativamente la integridad de los datos.
C√≥mo funciona: Se eliminan las filas o columnas que contienen valores nulos.
Consideraciones: Puede reducir el tama√±o del conjunto de datos y perder informaci√≥n valiosa si hay muchos valores nulos.
- Imputaci√≥n con estad√≠sticas descriptivas (media, mediana, moda):
Cu√°ndo usarlo: Es √∫til cuando queremos reemplazar los valores nulos con estimaciones basadas en la tendencia central de los datos.
C√≥mo funciona:
Media: Reemplaza los valores nulos con la media (promedio) de la columna.
Mediana: Reemplaza los valores nulos con la mediana (valor central) de la columna.
Moda: Reemplaza los valores nulos con la moda (valor m√°s frecuente) de la columna.
Consideraciones: Estas estrategias son robustas ante valores at√≠picos y preservan la distribuci√≥n general de los datos.
